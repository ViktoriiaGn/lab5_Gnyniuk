```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0, 10, 100)
epsilon=0.2
y=np.sin(x) + epsilon
plt.plot(x, y, label='sin(x)')
marker_list=["o", "^", "s", "D", "X"]
marker_index=range(0, len(x), 10)
for i , index in enumerate(marker_index):
  current_marker=marker_list[i%len(marker_list)]
  plt.scatter(x[index], y[index], marker=current_marker, color="red")
plt.title("Коливальний процес")
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
url = "https://api.open-meteo.com/v1/forecast?latitude=50.4547&longitude=30.5238&hourly=surface_pressure&timezone=auto&format=csv"

pressure = pd.read_csv(url, skiprows=3, parse_dates=['time'], index_col='time')
pressure.head(168).plot(figsize=(12, 6), color='orange', linewidth=1.5)

average=pressure.iloc[:,0].mean()
plt.axhline(y=average, color="red")
plt.title("Атмосферний тиск")
plt.ylabel("Тиск")
plt.xlabel("Дата")
plt.grid(True)

plt.show()
std=pressure.iloc[:, 0].std()
print(std)
```
